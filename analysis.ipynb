```python
import pandas as pd
import numpy as np
```

```python
df_trfc_lt = pd.read_csv('1.txt',header=None)
df_long_queue = pd.read_csv('2.txt',header=None)
df_static = pd.read_csv('3.txt',header=None)

df_trfc_lt['trfc_density']=df_trfc_lt[2]+df_trfc_lt[3]+df_trfc_lt[4]+df_trfc_lt[5]
df_trfc_lt['avg_tr_lt']=df_trfc_lt[0]/df_trfc_lt[1]
df_trfc_lt['avg_long_seq']=df_long_queue[0]/df_trfc_lt[1]
df_trfc_lt['avg_static']=df_static[0]/df_trfc_lt[1]

df_trfc_lt=df_trfc_lt.drop(columns=0)

df_trfc_lt['per_imp_tr_long']=((df_trfc_lt['avg_long_seq']-df_trfc_lt['avg_tr_lt'])/(df_trfc_lt['avg_long_seq']))*100
df_trfc_lt['per_imp_tr_static']=((df_trfc_lt['avg_static']-df_trfc_lt['avg_tr_lt'])/(df_trfc_lt['avg_static']))*100
```
